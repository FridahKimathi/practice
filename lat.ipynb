```python
# import necessary libraries
import pandas as pd
from openpyxl.styles import Border, Side, Alignment, PatternFill
```

```python
# main dataset
main_df = pd.read_csv("/content/main_dataset.csv")
# Preview the first 5 rows
main_df.head()
```

```python
# roster dataset
roster_df = pd.read_csv("/content/roster_dataset.csv")
# Preview the first 5 rows
main_df.head()
```

```python
# variable decription

variable_description = {
    "province_id": {"variable_type": "Numeric", "values": "N/A", "value_labels": "N/A", "variable_label": "Province code"},
    "province_name": {"variable_type": "String", "values": "N/A", "value_labels": "N/A", "variable_label": "Province of residence"},
    "district_id": {"variable_type": "Numeric", "values": "N/A", "value_labels": "N/A", "variable_label": "District code"},
    "district_name": {"variable_type": "String", "values": "N/A", "value_labels": "N/A", "variable_label": "District of residence"},
    "sector_id": {"variable_type": "Numeric", "values": "N/A", "value_labels": "N/A", "variable_label": "Sector code"},
    "sector_name": {"variable_type": "String", "values": "N/A", "value_labels": "N/A", "variable_label": "Sector of residence"},
    "cell_id": {"variable_type": "Numeric", "values": "N/A", "value_labels": "N/A", "variable_label": "Cell code"},
    "cell_name": {"variable_type": "String", "values": "N/A", "value_labels": "N/A", "variable_label": "Cell of residence"},
    "village_id": {"variable_type": "Numeric", "values": "N/A", "value_labels": "N/A", "variable_label": "Village code"},
    "village_name": {"variable_type": "String", "values": "N/A", "value_labels": "N/A", "variable_label": "Village of residence"},
    "list_available": {"variable_type": "Numeric", "values": [0, 1], "value_labels": ["No", "Yes"], "variable_label": "If the ubudehe list is available"},
    "list_source": {"variable_type": "Numeric", "values": [1, 2, 3], "value_labels": ["Sector Executive Officer", "Cell Executive Officer", "Village Leader"], "variable_label": "From whom the ubudehe list was obtained"},
    "ubudehe_contact": {"variable_type": "String", "values": "N/A", "value_labels": "N/A", "variable_label": "Ubudehe list contact person"},
    "contact_role": {"variable_type": "Numeric", "values": [1, 2, 3], "value_labels": ["Village leader", "Village in-charge of security", "Village Community Health Worker"], "variable_label": "Role of the ubudehe list contact person"},
    "contact_phone": {"variable_type": "Numeric", "values": "N/A", "value_labels": "N/A", "variable_label": "Phone number of ubudehe list contact person"},
    "num_households": {"variable_type": "Numeric", "values": "N/A", "value_labels": "N/A", "variable_label": "Number of households in ubudehe category 1"},
    "hh_head_position": {"variable_type": "Numeric", "values": "N/A", "value_labels": "N/A", "variable_label": "Position of household head on ubudehe list"},
    "name_hh_head": {"variable_type": "String", "values": "N/A", "value_labels": "N/A", "variable_label": "Name of the household head"},
    "nid_hh_head": {"variable_type": "String", "values": "N/A", "value_labels": "N/A", "variable_label": "National ID number of the household head"},
    "phone_hh_head": {"variable_type": "Numeric", "values": "N/A", "value_labels": "N/A", "variable_label": "Phone number of household head"},
    "gps": {"variable_type": "Numeric", "values": "N/A", "value_labels": "N/A", "variable_label": "GPS coordinates of where list was obtained"},
    "parent_key": {"variable_type": "String", "values": "NA", "value_labels": "NA", "variable_label": "Unique ID"}
}
```

**1 a)**

**GPS can be found in 4 columns that will be combined into one column named `GPS coordinates of where list was obtained`**

```python
# Combine GPS columns into a single 'GPS' column with a dictionary
main_df['GPS coordinates of where list was obtained'] = main_df.apply(lambda row:
                                                       f"latitude: {row['gpslatitude']}, longitude : {row['gpslongitude']}, altitude: {row['gpsaltitude']}, accuracy: {row['gpsaccuracy']}"
                                                       , axis=1)

# Drop the original GPS-related columns
main_df = main_df.drop(['gpslatitude', 'gpslongitude', 'gpsaltitude', 'gpsaccuracy'], axis=1)
# Preview the first 2 rows
main_df.head(2)
```

```python
# rename columns - From variable names to Variable labels
def rename_columns(df):
  df_updated_cols = [variable_description[col]["variable_label"] if col in variable_description.keys() else col for col in df.columns]

  df.columns = df_updated_cols
  return df
```

```python
# rename columns in main dataset
main_df = rename_columns(main_df)
# Preview the first 2 rows
main_df.head(2)
```

```python
# rename columns in roster dataset
roster_df = rename_columns(roster_df)
# Preview the first 2 rows
roster_df.head(2)
```

**1 b)**

```python
# Replace categorical variables with their values
replace_cat_columns = {
    variable_description[key]["variable_label"]: dict(zip(variable_description[key]["values"], variable_description[key]["value_labels"]))
    for key in variable_description.keys()
    if (key != "parent_key") & (variable_description[key]["value_labels"] != "N/A")
  }

# Replace categorical variables with their values in main dataset
main_df.replace(replace_cat_columns, inplace=True)

# Replace categorical variables with their values in roster dataset
roster_df.replace(replace_cat_columns, inplace=True)

# Preview the first 2 rows
main_df.head(2)
```

**2**

```python
# Location variables
location_variables = ['Province', 'District', 'Sector', 'Cell', 'Village']

# Create a single variable showing value label and embedded value
for location in location_variables:
    code_column = f'{location} code'
    residence_column = f'{location} of residence'
    combined_column = f'{location} combined'

    main_df[combined_column] = main_df.apply(
        lambda row: f"{row[code_column]} - {row[residence_column]}",
        axis=1)

# Preview the first 3 rows
main_df.head(3)
```

**3.**

```python
# combine main and roster datasets
merged_fridahkimathi_20240103 = main_df.merge(roster_df, on="Unique ID", how="inner")
# Preview the first 3 rows
merged_fridahkimathi_20240103.head(3)
```

**4 a) & b)**

```python
# List of Variable labels
variable_labels = [variable_description[key]["variable_label"] for key in variable_description.keys()]
```

```python
# Add combined location labels to the list in particular postions
variable_labels[2:2] = ["Province combined"]
variable_labels[5:5] = ["District combined"]
variable_labels[8:8] = ["Sector combined"]
variable_labels[11:11] = ["Cell combined"]
variable_labels[14:14] = ["Village combined"]
```

```python
# 4 a) Reorder the variables according to table 1
# 4 b) Rename the clean dataset
clean_fridahkimathi_20240103 = merged_fridahkimathi_20240103.reindex(columns=variable_labels)
# Preview the first 3 rows
clean_fridahkimathi_20240103.head(3)
```

**5.**

```python
# The unique districts in the datasets
districts = clean_fridahkimathi_20240103["District of residence"].unique()
districts

with pd.ExcelWriter("fridahkimathi-20240104.xlsx", engine="openpyxl") as writer:
  # Workbook
  workbook = writer.book

  for district in districts:
    district_df = clean_fridahkimathi_20240103[clean_fridahkimathi_20240103["District of residence"] == district]

    # Sectors in district
    sectors = clean_fridahkimathi_20240103["Sector of residence"].unique()

    for sector in sectors:
          sector_df = clean_fridahkimathi_20240103[clean_fridahkimathi_20240103["Sector of residence"] == sector]

          # Write the sector data to Excel
          sector_df.to_excel(writer, sheet_name=f"{district}-{sector}", index=False, freeze_panes=(1,0))

          # worksheet
          worksheet = workbook[f"{district}-{sector}"]

          # Autofit the excel file columns
          for column in worksheet.columns:
              max_length = max(len(str(cell.value)) for cell in column)
              adjusted_width = (max_length + 1) * 0.98
              worksheet.column_dimensions[column[0].column_letter].width = adjusted_width

          # Add all borders
          for row in worksheet.iter_rows(min_row=1, max_row=worksheet.max_row, min_col=1, max_col=worksheet.max_column):
              for cell in row:
                  cell.border = Border(left = Side(border_style="thin"), right = Side(border_style="thin"),
                                        top = Side(border_style="thin"), bottom = Side(border_style="thin"))
          # Add background to the column titles
          for row in worksheet.iter_rows(min_row=1, max_row=1, min_col=1, max_col=worksheet.max_column):
              for cell in row:
                      cell.fill = PatternFill(start_color='92CDDC', end_color='92CDDC', fill_type='solid')
```
